```python
from plotnine import *
from plotnine.data import anscombe_quartet, mtcars
from great_tables import GT
from great_tables.data import sp500
import pandas as pd
```

```python
(
    ggplot(anscombe_quartet, aes(x='x', y='y')) 
    + geom_point(color='sienna',fill='darkorange', size=2)
    + geom_smooth(method='lm', se=False, fullrange=True, color='darkblue')
    + facet_wrap('dataset')
    + scale_y_continuous(breaks=(4, 8, 12))
    + theme_minimal(11)
    + coord_fixed(ylim=(2,14), xlim=(3,22))
    + labs(title='Anscombe Quartet')
    + theme(axis_line=element_line(color='#000000'),            
            axis_title=element_blank(),
            axis_ticks=element_blank(),
            panel_spacing=.04     
    )
)
```

```python
df = pd.DataFrame(mtcars)
hp_wt_df = df[['hp', 'wt']]
(
    ggplot(hp_wt_df, aes(x='wt',y='hp'))
    + geom_point()
    + theme_tufte()
    + geom_smooth(method='ols', se=False)
    + labs(x='Weight',y='Horsepower')
)
```

```python
df = pd.DataFrame(mtcars)
(
    ggplot(df, aes(x='factor(cyl)', fill='factor(cyl)'))
    + geom_bar()
    + geom_text(aes(label=after_stat('prop*100'), group=1), 
                stat='count', 
                nudge_y=(0.2),
                format_string="{:.1f}%"
    )
    + facet_wrap("gear")
    + coord_fixed(.5)
)
```

```python
stock_df = sp500

stock_df = stock_df[(stock_df["date"] >= "2015-01-01") & (stock_df["date"] <= "2015-12-31")]
stock_df = stock_df.sort_values(by='date')
(
    GT(stock_df)
    .tab_header(title='S&P 500 Performance', subtitle='Jan 1, 2015 - Dec 31, 2015')
)
```
