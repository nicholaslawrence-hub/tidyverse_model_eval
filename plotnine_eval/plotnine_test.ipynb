```python
life_df = pd.read_csv('../example_datasets/Social_media_impact_on_life.csv')
life_df = life_df[life_df['Most_Used_Platform'].isin(['Instagram','Facebook','Youtube','Twitter','WhatsApp'])]

def darken(hex_colors:str, f:str) -> str:
    return;
(
ggplot(life_df, aes(x='Most_Used_Platform', y='Avg_Daily_Usage_Hours', color='Gender'))
    #+ geom_col(aes(fill='Most_Used_Platform'), stat='summary', fun_y=np.mean, alpha=0.5) 
    + geom_segment(
        aes(
            x=stage('Most_Used_Platform', after_stat='x-.45'),
            xend=stage('Most_Used_Platform', after_stat='x+.45'),
            yend=after_stat('y')
        ),
        stat = "summary",
        fun_y = np.mean,
        color='black',
        size=1.2
    )
    + geom_segment(
        aes(
            x=stage('Most_Used_Platform', after_stat='x-.45'),
            xend=stage('Most_Used_Platform', after_stat='x+.45'),
            yend=after_stat('y')
        ),
        stat = "summary",
        fun_y = np.mean,
        position = position_dodge(width=0.9),
        size=1.2
    )
    + geom_text(
        aes(y=stage('Avg_Daily_Usage_Hours', after_stat="ymin-.2"),
        label=after_stat("n")
    ),
    stat = 'summary',
    position = position_dodge(width=0.9),
    fun_ymin = np.min, 
    format_string = "n = {}",
    )
    + geom_sina()
    + labs(x='Most Used Platform', y='Daily Use Hours')
)
```
