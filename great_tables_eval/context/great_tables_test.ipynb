```python
import pandas as pd
from great_tables import GT, html, style, loc, google_font
from great_tables.data import sp500
```

```python
caltrain_df = pd.read_csv("../../example_datasets/caltrain_weekend_northbound.csv")
caltrain_df.insert(0, 'fare_zone', [4]*4 + [3]*6 + [2]*10 + [1]*4)
caltrain_df.head(5)
```

```python
train_columns=caltrain_df.filter(regex=("^6")).columns.tolist()
caltrain_table = (
    GT(caltrain_df)
    .tab_header(
        title=html(
            """
                <span style="color: rgb(224,37,33); font-weight: bold; font-size: 40px;">Northbound &ndash;</span>
                <span style="color: black; font-weight: bold; font-size: 30px">
                    WEEKEND SERVICE to SAN FRANCISCO
                </span>
            """
        )
    )
    .tab_stub('fare_zone')
    .tab_source_note(
        source_note=
        """
            Sourced from CalTrain's 01/31/26 weekend schedule. 
        """
    )
    .cols_width(
        cases={
            'Stop':'300px'
        }
    )
    .opt_align_table_header(align='left')
    .cols_align(align='center')
    .tab_style(
        locations=loc.header(),
        style=style.css(
            "color: rgb(224,37,33) !important; font-size: 35px !important; font-weight: bold !important; border-width: 0px !important;",
        )
    )
    .tab_style(
        locations=loc.column_labels(),
        style=style.css(
            "background-color: black !important; color: white !important; font-size: 16px !important; font-weight: bold !important; border-left: grey 4px !important; border-width: 4px !important; border-right: grey 4px !important;"
        )
    )
    .tab_style(
        locations=loc.stub(),
        style=style.css(
            "background-color: rgb(224,37,33) !important; color: white !important; writing-mode: sideways-lr; font-size: 14px !important; font-weight: normal !important; align: left !important"
        )
    )
    .tab_style(
        locations=loc.body(),
        style=style.css(
            """
            height: 10px !important; 
            border-left: solid #a2a7a8 1px !important;
            border-right: solid #a2a7a8 1px !important;
            color: black !important
            """
        )
    )
    .tab_options(
        row_group_as_column=True,
        row_striping_background_color="#e1e2e2"
    )
    .cols_label(
        Stop='Train No.',
        fare_zone='Fare Zone'
    )
    .opt_row_striping()
    .opt_table_outline()
    .opt_table_font(font=google_font('frutiger'))
)
caltrain_table
```
